```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

cleaneddata_movies = pd.read_csv('cleaneddata_all_movies.csv')
```

cleaneddata_movies

```python
# ROI is calculated as : Profit / Budget
# Profit is calculated as : BoxOffice Collections - Budget

cleaneddata_movies['Profit (Crores)'] = cleaneddata_movies['BoxOffice (Crores)'] - cleaneddata_movies['Budget (Crores)']
cleaneddata_movies
```

```python
cleaneddata_movies['ROI'] = (cleaneddata_movies['Profit (Crores)'] / cleaneddata_movies['Budget (Crores)']).round(2)
cleaneddata_movies
```

# Calculate ROI (Return on Investment) for each movie

```python
ROI = cleaneddata_movies[['MovieID','Title','Language','Budget (Crores)','BoxOffice (Crores)','Profit (Crores)','ROI']]
ROI
```

# Which are the top 3 movies with highest ROI across all industries?


```python
TOP_3_MOVIES = ROI.sort_values(by='ROI', ascending=False).head(3)
TOP_3_MOVIES
```

# Which are the top 3 Hindi movies with highest ROI?


```python
language = 'Hindi'
TOP_3_HINDI = ROI[ROI['Language'].str.lower()==language.lower()]\
.sort_values(by='ROI',ascending=False)\
.head(3)
TOP_3_HINDI
```

# Which are the top 3 Telugu movies with highest ROI?


```python
language = 'TelUgu'
TOP_3_TELEGU = ROI[ROI['Language'].str.lower()==language.lower()]\
.sort_values(by='ROI',ascending=False)\
.head(3)
TOP_3_TELEGU
```

# Plot all this information in separate graphs


```python
# Topmost 3 for hindi
plt.figure(figsize=(6, 6))
bars = plt.bar(TOP_3_HINDI['Title'], TOP_3_HINDI['ROI'], color='lightcoral')
plt.title('Top 3 Hindi Movies with Highest ROI')
plt.xlabel('Title')
plt.ylabel('ROI')
plt.xticks(rotation=45, ha='right')

# Add ROI values on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.2f}', 
             ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# TOPMOST 3 FOR telegu
plt.figure(figsize=(6, 6))
bars = plt.bar(TOP_3_TELEGU['Title'], TOP_3_TELEGU['ROI'], color='lightcoral')
plt.title('Top 3 TELEGU Movies with Highest ROI')
plt.xlabel('Title')
plt.ylabel('ROI')
plt.xticks(rotation=45, ha='right')

# Add ROI values on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.2f}', 
             ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Topmost 3 for  all movies
plt.figure(figsize=(9, 6))
bars = plt.barh(TOP_3_MOVIES['Title'], TOP_3_MOVIES['ROI'], color='skyblue')
plt.title('Top 3 Movies with Highest ROI across all industries')
plt.xlabel('ROI')
plt.ylabel('Title')

# Add ROI values on top of bars
for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.05, bar.get_y() + bar.get_height() / 2, 
             f'{width:.2f}', va='center', fontsize=10)

plt.tight_layout()
plt.show()
```
